```python
import os
import glob
from rembg import remove
from PIL import Image

root = os.listdir('/data/ephemeral/home/origin_data/train/images/')
for person in root:
    imgs = glob.glob('/data/ephemeral/home/origin_data/train/images/'+person +'/*.jpg')
    if person.startswith('.'):
        continue
    target_root = '/data/ephemeral/home/removed_background/'
    if not os.path.isdir(target_root+'/'+person):
        os.mkdir(target_root+'/'+person)
    for img in imgs:
        input_img = img
        target_img = target_root+person+'/' +img.split('/')[-1].split('.')[0] +'.png'
        image = Image.open(input_img)
        # image = Image.convert('RGB')
        output = remove(image)
        output.save(target_img)
    print(person)
print("DONE")
```
